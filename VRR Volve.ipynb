```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel("VRR_VOLVE_Input.xlsx")
df
```

```python
w=df["PROD-Gas"]
x=df["PROD-Oil"]
y=df["PROD-Water"]
z=df["Wi-VOLUME"]

df["CUM.gas"]=np.cumsum(w)
df["CUM.oil"]=np.cumsum(x)
df["CUM.water"]=np.cumsum(y)
df["CUM.waterinj"]=np.cumsum(z)
df
```

```python
#Instantaneous_VRR
df["VRR"]=z/((x*df["Bo"])+(y*df["Bw"])+(df["Bg"]*x*(df["GOR"]-df["Rs"])))
#Cumulative_VRR
df["CUM.VRR"]=df["CUM.waterinj"]/((df["CUM.oil"]*df["Bo"])+(df["CUM.water"]*df["Bw"])+(df["Bg"]*df["CUM.oil"]*(df["GOR"]-df["Rs"])))
#Cumulative_VRR without FVF
df['CUM_VRR_without_FVF']=df["CUM.waterinj"]/((df["CUM.oil"])+(df["CUM.water"])+(df["CUM.gas"]))
df
```

```python
fig,ax1=plt.subplots()
ax1.plot(df["Date"],df["VRR"],color="blue",label="Instantaneous_VRR")
ax1.set_xlabel("YEAR")
ax1.set_ylabel("Instantaneous_VRR")
ax1.tick_params(axis='y')
ax2=ax1.twinx()
ax2.plot(df["Date"],df["CUM.VRR"],color="green",label="Cummulative_VRR")
ax2.set_ylabel("Cummulative_VRR")
ax2.tick_params(axis='y')
plt.title('Instantaneous_VRR vs Cummulative_VRR')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="lower right")

fig.tight_layout()
```

```python
df1=df.copy()
C=df1["VRR"]
df1["VRR___CUM."]=np.cumsum(C)
df1
```

```python
fig,ax1=plt.subplots()
ax1.plot(df1["Date"],df1["VRR"],color="blue",label="Instantaneous_VRR")
ax1.set_xlabel("YEAR")
ax1.set_ylabel("Instantaneous_VRR")
ax1.tick_params(axis='y')
ax2=ax1.twinx()
ax2.plot(df1["Date"],df1["CUM_VRR_without_FVF"],color="brown",label="Cummulative_VRR")
ax2.set_ylabel("Cummulative_VRR")
ax2.tick_params(axis='y')
plt.title('Instantaneous_VRR vs Cummulative_VRR')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="best")

fig.tight_layout()
```
